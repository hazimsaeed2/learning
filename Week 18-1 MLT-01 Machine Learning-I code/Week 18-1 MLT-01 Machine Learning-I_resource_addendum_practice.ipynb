# MLT-01 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for ML workflow, leakage controls, Gini impurity, and local fallback scope.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['mlt01_source_pdf_inventory.csv','mlt01_source_zip_inventory.csv','mlt01_data_inventory.csv','mlt01_feature_engineering_audit.csv','mlt01_gini_impurity_worked_example.csv','mlt01_split_leakage_controls.csv','mlt01_model_validation_metrics.csv','mlt01_sklearn_fallback_scope.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and data inventory

The local data files match the lecture workflow: TSLA OHLCV data and the engineered ML dataset.

```python
print(data['mlt01_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['mlt01_source_zip_inventory.csv'].groupby('role').size().to_string())
print(data['mlt01_data_inventory.csv'].to_string(index=False))
assert data['mlt01_data_inventory.csv']['file'].str.contains('TSLA').any()
```
