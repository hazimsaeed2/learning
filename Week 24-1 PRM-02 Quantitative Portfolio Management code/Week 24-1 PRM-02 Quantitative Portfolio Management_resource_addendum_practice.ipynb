# PRM-02 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for portfolio return/risk, diversification, frontier optimization, out-of-sample testing, and risk-adjusted ratios.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['prm02_source_pdf_inventory.csv', 'prm02_source_zip_inventory.csv', 'prm02_data_inventory.csv', 'prm02_return_risk_metrics.csv', 'prm02_diversification_metrics.csv', 'prm02_frontier_metrics.csv', 'prm02_frontier_weights.csv', 'prm02_out_of_sample_metrics.csv', 'prm02_risk_adjusted_metrics.csv', 'prm02_treynor_correction.csv', 'prm02_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and data inventory

```python
print(data['prm02_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['prm02_source_zip_inventory.csv'].groupby(['zip','role']).size().to_string())
print(data['prm02_data_inventory.csv'][['file','rows','columns','date_start','date_end','missing_cells']].to_string(index=False))
assert data['prm02_data_inventory.csv'].query("file == 'nifty50.csv'")['rows'].iloc[0] == 1393
```

## 2. Return, risk, and diversification checks

```python
print(data['prm02_return_risk_metrics.csv'].to_string(index=False))
print(data['prm02_diversification_metrics.csv'].to_string(index=False))
tcs_vol = data['prm02_return_risk_metrics.csv'].query("asset == 'TCS'")['annual_volatility_pct'].iloc[0]
pair_vol = data['prm02_diversification_metrics.csv'].query("case == 'TCS+MARICO_50_50'")['annual_volatility_pct'].iloc[0]
assert pair_vol < tcs_vol
```

## 3. Frontier, out-of-sample, and ratio controls

```python
print(data['prm02_frontier_metrics.csv'].to_string(index=False))
print(data['prm02_out_of_sample_metrics.csv'].to_string(index=False))
print(data['prm02_risk_adjusted_metrics.csv'].to_string(index=False))
print(data['prm02_treynor_correction.csv'].to_string(index=False))
corr = data['prm02_treynor_correction.csv'].iloc[0]
assert corr['corrected_annual_value'] > corr['original_notebook_value']
```
