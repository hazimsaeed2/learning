# TBP-01 IB Python API validation addendum

This additive notebook validates the mock request/response, async timing, callback flow, source materials, and live-scope controls. It does not connect to Interactive Brokers.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['tbp01_html_audit.csv', 'tbp01_source_inventory.csv', 'tbp01_notebook_execution.csv', 'tbp01_mock_api_metrics.csv', 'tbp01_historical_data_checks.csv', 'tbp01_live_scope_controls.csv', 'tbp01_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and execution audit

```python
print(data['tbp01_source_inventory.csv'].to_string(index=False))
print(data['tbp01_notebook_execution.csv'].to_string(index=False))
nb = data['tbp01_notebook_execution.csv'].set_index('metric')
assert int(nb.loc['safe_execution_errors','value']) == 0
```

## 2. Mock API and historical-data checks

```python
print(data['tbp01_mock_api_metrics.csv'].to_string(index=False))
print(data['tbp01_historical_data_checks.csv'].to_string(index=False))
h = data['tbp01_historical_data_checks.csv'].set_index('metric')
assert int(h.loc['callback_rows_requested','value']) == 10
```

## 3. Scope and validation controls

```python
print(data['tbp01_live_scope_controls.csv'].to_string(index=False))
print(data['tbp01_validation_controls.csv'].to_string(index=False))
scope = data['tbp01_live_scope_controls.csv']
assert (scope['validated_here'] == 'no').sum() >= 4
```
