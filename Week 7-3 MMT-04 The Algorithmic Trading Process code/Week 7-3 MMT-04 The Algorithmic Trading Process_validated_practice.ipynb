# MMT-04 validated additive notebook

There is no original practice notebook in the root lecture folder; this notebook is the validated additive execution layer.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['mmt04_source_pdf_inventory.csv', 'mmt04_source_zip_inventory.csv', 'mmt04_tca_example_inputs.csv', 'mmt04_implementation_shortfall_breakdown.csv', 'mmt04_post_trade_metrics.csv', 'mmt04_pre_trade_model_parameters.csv', 'mmt04_mi_data_diagnostics.csv', 'mmt04_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source inventory

```python
print(data['mmt04_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['mmt04_source_zip_inventory.csv'].groupby(['zip','role']).size().to_string())
assert data['mmt04_source_zip_inventory.csv'].query("role == 'source_notebook'").shape[0] >= 8
```

## 2. TCA identity and post-trade checks

```python
print(data['mmt04_tca_example_inputs.csv'].to_string(index=False))
print(data['mmt04_implementation_shortfall_breakdown.csv'].to_string(index=False))
print(data['mmt04_post_trade_metrics.csv'].to_string(index=False))
b = data['mmt04_implementation_shortfall_breakdown.csv'].set_index('component')
component_sum = b.loc[['Delay cost','Execution cost','Opportunity cost','Fixed cost'], 'value_usd'].sum()
assert component_sum == b.loc['Implementation shortfall','value_usd']
```

## 3. MI data and pre-trade model checks

```python
print(data['mmt04_mi_data_diagnostics.csv'].to_string(index=False))
print(data['mmt04_pre_trade_model_parameters.csv'].to_string(index=False))
print(data['mmt04_validation_controls.csv'].to_string(index=False))
diag = data['mmt04_mi_data_diagnostics.csv'].set_index('field')
assert diag.loc['Cost','rows'] == 233135
```
