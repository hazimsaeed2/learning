```python
# Step 5. User Defined TCA Functions

# Implementation Shortfall
def IS_Cost_calc(Pd, P0, Pn, Pavg, S, X, R, comm, Side):
    Delay=S*(P0-Pd)*Side
    Exec=X*(Pavg-P0)*Side
    Oppor=R*(Pn-P0)*Side
    Fixed=X*comm
    ImplShortFall=Delay+Exec+Oppor+Fixed
    return ImplShortFall, Delay, Exec, Oppor, Fixed

# Arrival Cost
def Arrival_Cost_calc(P0, Pavg, Side):
    arrival_cost = Side * (Pavg-P0)/P0 * 10**4
    return arrival_cost

# VWAP Slippage
def VWAP_Slippage_calc(VWAP, Pavg, Side):
    vwap_slippage = Side * (Pavg-VWAP)/VWAP * 10**4
    return vwap_slippage


# Market Impact
def MI_calc(Size, Volatility, POV, MIParams):
    a1, a2, a3, a4, b1 = MIParams
    mi = (a1*Size**a2*Volatility**a3) * (b1*POV**a4 + (1-b1))
    return mi

# Timing Risk
def TR_calc(Size, Volatility, POV):
    tr = Volatility * ((1/3*1/250*Size*(1-POV)/POV )**0.5) * 10**4
    return tr

# Value-Add Calculation
def ValueAdd_calc(ArrivalCost, MI, TR):
    va = MI-ArrivalCost
    z= (MI-ArrivalCost)/TR
    return va, z
```
