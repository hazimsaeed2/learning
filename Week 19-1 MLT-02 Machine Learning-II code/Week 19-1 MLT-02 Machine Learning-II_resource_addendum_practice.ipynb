# MLT-02 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for SVM metrics, confusion matrices, K-Means clustering, and model-risk controls.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['mlt02_source_pdf_inventory.csv','mlt02_source_zip_inventory.csv','mlt02_data_inventory.csv','mlt02_target_feature_leakage_audit.csv','mlt02_svm_kernel_metrics.csv','mlt02_confusion_metrics.csv','mlt02_kmeans_cluster_summary.csv','mlt02_elbow_diagnostics.csv','mlt02_bias_variance_controls.csv','mlt02_sklearn_fallback_scope.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and data inventory

The local packet includes TSLA, company fundamentals, Nifty-50 constituents, and the original source notebooks.

```python
print(data['mlt02_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['mlt02_source_zip_inventory.csv'].groupby('role').size().to_string())
print(data['mlt02_data_inventory.csv'].to_string(index=False))
assert data['mlt02_data_inventory.csv']['file'].str.contains('companydata').any()
```

## 2. SVM metrics and leakage controls

The target is next-open direction and volume is lagged first, which is a stronger anti-leakage control than using same-day volume.

```python
print(data['mlt02_target_feature_leakage_audit.csv'].to_string(index=False))
print(data['mlt02_svm_kernel_metrics.csv'].to_string(index=False))
print(data['mlt02_confusion_metrics.csv'].to_string(index=False))
assert data['mlt02_target_feature_leakage_audit.csv']['item'].str.contains('Volume').any()
```

## 3. K-Means and model-risk controls

K-Means inertia always falls as k rises; the elbow is a judgement and should be checked against business interpretability.

```python
print(data['mlt02_kmeans_cluster_summary.csv'].to_string(index=False))
print(data['mlt02_elbow_diagnostics.csv'].to_string(index=False))
print(data['mlt02_bias_variance_controls.csv'].to_string(index=False))
print(data['mlt02_sklearn_fallback_scope.csv'].to_string(index=False))
assert data['mlt02_bias_variance_controls.csv']['risk'].str.contains('KMeans').any()
```
