# TBP-05 resource addendum

This additive notebook preserves the original notebook and adds dependency-safe validation tables for backtest and live-trading controls.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['tbp05_source_pdf_inventory.csv','tbp05_source_zip_inventory.csv','tbp05_existing_code_inventory.csv','tbp05_model_validation_metrics.csv','tbp05_backtest_trap_controls.csv','tbp05_ibridgepy_reference_map.csv','tbp05_live_robot_control_matrix.csv','tbp05_dependency_fallbacks.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and code inventory

The lecture uses PDFs, a zip packet, SPY.csv, existing scripts, and IBridgePy reference robots.

```python
print(data['tbp05_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['tbp05_source_zip_inventory.csv'].groupby('role').size().to_string())
print(data['tbp05_existing_code_inventory.csv'][['file','role']].head(20).to_string(index=False))
assert data['tbp05_source_zip_inventory.csv']['role'].str.contains('script').any()
assert data['tbp05_existing_code_inventory.csv']['file'].str.contains('SPY.csv').any()
```

## 2. Model validation metrics

The NumPy fallback reproduces the core regression and backtest checks without scikit-learn.

```python
metrics = data['tbp05_model_validation_metrics.csv']
print(metrics.to_string(index=False))
assert float(metrics.query("metric == 'ols_slope'")['value'].iloc[0]) < 0
assert metrics['metric'].str.contains('test_strategy_final').any()
```

## 3. Backtest traps and live robot controls

Cost, overfit, and production controls are separate from the signal idea.

```python
traps = data['tbp05_backtest_trap_controls.csv']
live = data['tbp05_live_robot_control_matrix.csv']
print(traps.to_string(index=False))
print(live.to_string(index=False))
assert traps['trap'].str.contains('Transaction').any()
assert live['control'].str.contains('Early close').any()
```

## 4. IBridgePy reference map and dependency fallbacks

The reference robots are not standalone scripts; they require IBridgePy and an Interactive Brokers connection.

```python
ib = data['tbp05_ibridgepy_reference_map.csv']
deps = data['tbp05_dependency_fallbacks.csv']
print(ib.to_string(index=False))
print(deps.to_string(index=False))
assert ib['file'].str.contains('handle_error').any()
assert deps['additive_fix'].str.contains('LinearRegression').any()
```
