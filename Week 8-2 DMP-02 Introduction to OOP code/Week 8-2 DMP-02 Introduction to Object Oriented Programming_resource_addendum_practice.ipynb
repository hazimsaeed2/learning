# DMP-02 OOP validation addendum

This additive notebook validates the shipped OOP practice notebook, core class/object concepts, and the BacktestingCrossover strategy class. It does not change the original notebook.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['dmp02_html_audit.csv', 'dmp02_source_inventory.csv', 'dmp02_notebook_execution.csv', 'dmp02_oop_concept_checks.csv', 'dmp02_backtest_metrics.csv', 'dmp02_class_design_controls.csv', 'dmp02_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and notebook audit

```python
print(data['dmp02_source_inventory.csv'].to_string(index=False))
print(data['dmp02_notebook_execution.csv'].to_string(index=False))
nb = data['dmp02_notebook_execution.csv'].set_index('metric')
assert int(nb.loc['original_execution_errors','value']) == 0
```

## 2. OOP concept checks

```python
print(data['dmp02_oop_concept_checks.csv'].to_string(index=False))
checks = data['dmp02_oop_concept_checks.csv'].set_index('check')
assert checks.loc['integer_dunder_add','value'] in [4, '4']
assert checks.loc['sporty_is_student','value'] in [True, 'True']
```

## 3. Backtest and class design controls

```python
print(data['dmp02_backtest_metrics.csv'].to_string(index=False))
print(data['dmp02_class_design_controls.csv'].to_string(index=False))
backtests = data['dmp02_backtest_metrics.csv'].set_index('run')
assert int(backtests.loc['SMA10_20','rows']) == 942
assert 'positions()' in backtests.loc['SMA10_20','position_lag_location']
```

## 4. Validation controls

```python
print(data['dmp02_validation_controls.csv'].to_string(index=False))
assert data['dmp02_validation_controls.csv'].shape[0] >= 6
```
