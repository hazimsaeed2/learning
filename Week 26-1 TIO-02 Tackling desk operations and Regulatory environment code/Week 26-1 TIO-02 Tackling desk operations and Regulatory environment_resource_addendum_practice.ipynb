# TIO-02 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for desk operations, latency, membership economics, co-location, network routing, operational risk, and current regulatory updates.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['tio02_source_pdf_inventory.csv', 'tio02_dependency_scope.csv', 'tio02_latency_paradigm_table.csv', 'tio02_entity_funds_table.csv', 'tio02_membership_cost_metrics.csv', 'tio02_colocation_network_metrics.csv', 'tio02_line_latency_cost_table.csv', 'tio02_platform_ladder.csv', 'tio02_setup_capex_plan.csv', 'tio02_operational_risk_controls.csv', 'tio02_current_regulatory_update.csv', 'tio02_unit_corrections.csv', 'tio02_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source, dependency, and regulatory checks

```python
print(data['tio02_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['tio02_dependency_scope.csv'].to_string(index=False))
print(data['tio02_current_regulatory_update.csv'].to_string(index=False))
assert 'FINRA Notice 26-10' in data['tio02_current_regulatory_update.csv']['current_update'].iloc[0]
```

## 2. Latency, membership, and infrastructure checks

```python
print(data['tio02_latency_paradigm_table.csv'].to_string(index=False))
print(data['tio02_membership_cost_metrics.csv'].to_string(index=False))
print(data['tio02_colocation_network_metrics.csv'].to_string(index=False))
m = data['tio02_membership_cost_metrics.csv'].set_index('metric')
assert m.loc['membership_breakeven_million_orders_per_year','value'] > 2000
```

## 3. Platform, capex, and operational controls

```python
print(data['tio02_line_latency_cost_table.csv'].to_string(index=False))
print(data['tio02_platform_ladder.csv'].to_string(index=False))
print(data['tio02_setup_capex_plan.csv'].tail(1).to_string(index=False))
print(data['tio02_operational_risk_controls.csv'].to_string(index=False))
print(data['tio02_unit_corrections.csv'].to_string(index=False))
ladder = data['tio02_platform_ladder.csv'].set_index('platform')
assert ladder.loc['FPGA / ASIC','x_faster_than_broker'] == 60000
```
