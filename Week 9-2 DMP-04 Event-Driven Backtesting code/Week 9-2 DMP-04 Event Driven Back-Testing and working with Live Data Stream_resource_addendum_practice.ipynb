# DMP-04 event-driven validation addendum

This additive notebook validates the shipped event-driven practice notebook, event contracts, event counts, parameter sweep, and risk statistics. It does not change the original notebook.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['dmp04_html_audit.csv', 'dmp04_source_inventory.csv', 'dmp04_notebook_execution.csv', 'dmp04_event_contract_checks.csv', 'dmp04_backtest_sweep_metrics.csv', 'dmp04_risk_stats.csv', 'dmp04_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and notebook audit

```python
print(data['dmp04_source_inventory.csv'].to_string(index=False))
print(data['dmp04_notebook_execution.csv'].to_string(index=False))
nb = data['dmp04_notebook_execution.csv'].set_index('metric')
assert int(nb.loc['original_execution_errors','value']) == 0
```

## 2. Event contract checks

```python
print(data['dmp04_event_contract_checks.csv'].to_string(index=False))
ev = data['dmp04_event_contract_checks.csv'].set_index('check')
assert int(ev.loc['market_events','value']) == 962
assert ev.loc['queue_type','value'] == 'collections.deque'
```

## 3. Parameter sweep and risk checks

```python
print(data['dmp04_backtest_sweep_metrics.csv'].to_string(index=False))
print(data['dmp04_risk_stats.csv'].to_string(index=False))
sweep = data['dmp04_backtest_sweep_metrics.csv']
assert sweep.shape[0] == 12
assert sweep['total_return_pct'].max() > 80
```

## 4. Validation controls

```python
print(data['dmp04_validation_controls.csv'].to_string(index=False))
assert data['dmp04_validation_controls.csv'].shape[0] >= 6
```
