## Troubleshooting

### Common Setup Issues

**Issue**: `uv: command not found`  
**Solution**: Add UV to PATH:
```bash
export PATH="$HOME/.cargo/bin:$PATH"
# Or restart terminal after UV install
```

**Issue**: `yfinance` download fails (rate limiting)  
**Solution**: Use cached data:
```python
df = pd.read_parquet('../data/raw/XLF.parquet')
```

**Issue**: `ModuleNotFoundError: No module named 'mlt04'`  
**Solution**: Install package from repo root:
```bash
cd quantinsti_mlt04  # Repo root
uv pip install -e .
```

**Issue**: Jupyter kernel crashes  
**Solution**: Restart kernel:
- **Menu**: Kernel → Restart Kernel
- **Shortcut**: Press `0 0` (zero twice) in Jupyter Lab
