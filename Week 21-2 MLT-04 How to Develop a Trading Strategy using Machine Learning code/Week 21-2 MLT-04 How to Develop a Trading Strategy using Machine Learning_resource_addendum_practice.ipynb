# MLT-04 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for the XLF ML strategy-development workflow.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['mlt04_source_pdf_inventory.csv','mlt04_source_zip_summary.csv','mlt04_local_file_inventory.csv','mlt04_parquet_data_inventory.csv','mlt04_feature_family_audit.csv','mlt04_saved_model_results.csv','mlt04_batch_comparison.csv','mlt04_holdout_results.csv','mlt04_leakage_controls.csv','mlt04_validation_controls.csv','mlt04_dependency_fallback_scope.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

The local parquet files cannot be read in this environment without pyarrow or fastparquet, so this table preserves the notebook's printed dimensions and dates.
