# OTS-01 resource addendum

This additive notebook preserves the original practice notebook and adds source-backed checks for compliance workflow, observation-to-business conversion, edge/risk separation, trade review, and psychology controls.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = {
    'sources': 'ots01_source_pdf_inventory.csv',
    'compliance': 'ots01_compliance_workflow.csv',
    'pipeline': 'ots01_observation_to_business_pipeline.csv',
    'edge_risk': 'ots01_edge_risk_map.csv',
    'review': 'ots01_trade_review_noise_demo.csv',
    'controls': 'ots01_psychology_automation_controls.csv',
}
data = {k: pd.read_csv(base / v) for k, v in files.items()}
{k: df.shape for k, df in data.items()}
```

## 1. Compliance and controls

The lecture opening and source notes emphasize that options examples are educational, market rules differ, and new strategies or order workflow changes need written approval.

```python
compliance = data['compliance']
print(compliance[['stage', 'student_action']].to_string(index=False))
assert 'Written approval' in set(compliance['stage'])
assert compliance['stage'].str.contains('Independent risk').any()
```

## 2. Observation first, method second

The source packet repeats the key process rule: do not start from a model. Start from a phenomenon, quantify it simply, then prove it can survive implementation.

```python
pipeline = data['pipeline'].sort_values('step')
print(pipeline[['step', 'stage', 'lecture_anchor']].to_string(index=False))
assert pipeline.iloc[0]['stage'] == 'Observation'
assert pipeline.iloc[-1]['stage'] == 'Review'
```

## 3. Edge/risk separation

Different participants can have different edges in the same instrument. The job is to name the edge and treat every other exposure as risk.

```python
edge_risk = data['edge_risk']
print(edge_risk[['participant', 'possible_edge', 'main_risks']].to_string(index=False))
assert edge_risk['main_risks'].str.contains('adverse selection|jump risk|overtrading', case=False, regex=True).any()
```

## 4. Trade review without narrating noise

The live-month sample below is intentionally short. It can make a weekday story feel persuasive even though the original evidence used 1000 observations.

```python
review = data['review']
summary = review.groupby(['phase', 'is_thursday'])['pnl'].agg(['count', 'mean']).round(4)
print(summary)
back_n = (review['phase'] == 'backtest_1000_obs').sum()
live_n = (review['phase'] == 'live_20_obs').sum()
assert back_n == 1000 and live_n == 20
```

## 5. Psychology and automation controls

These controls are not edge. They keep a real edge from being destroyed by overtrading, unnecessary automation, impossible goals, or P&L fixation.

```python
controls = data['controls']
print(controls[['control_area', 'lecture_rule']].to_string(index=False))
assert controls['lecture_rule'].str.contains('Automate needs').any()
assert controls['lecture_rule'].str.contains('Monitor the edge').any()
```
