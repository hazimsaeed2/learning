# MLT-03 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for neural-network features, gradient behavior, backtesting, TF-IDF, and local fallback scope.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['mlt03_source_pdf_inventory.csv','mlt03_source_zip_inventory.csv','mlt03_guided_project_inventory.csv','mlt03_data_inventory.csv','mlt03_feature_indicator_audit.csv','mlt03_activation_gradient_reference.csv','mlt03_mlp_validation_metrics.csv','mlt03_backtest_performance.csv','mlt03_tfidf_reference.csv','mlt03_model_risk_controls.csv','mlt03_fallback_scope.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Sources, data, and features

The local packet includes EUR/USD train and backtest data plus the neural-network source notebooks.

```python
print(data['mlt03_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['mlt03_source_zip_inventory.csv'].groupby('role').size().to_string())
print(data['mlt03_data_inventory.csv'].to_string(index=False))
print(data['mlt03_feature_indicator_audit.csv'].head(20).to_string(index=False))
assert data['mlt03_data_inventory.csv']['file'].str.contains('EURUSD').any()
```
