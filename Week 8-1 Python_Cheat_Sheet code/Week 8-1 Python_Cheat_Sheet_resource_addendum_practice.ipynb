# Python cheat-sheet resource addendum

This additive notebook validates the static cheat sheet with offline checks only. It does not fetch live market data.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['pycs_html_audit.csv', 'pycs_dependency_scope.csv', 'pycs_parameter_corrections.csv', 'pycs_workflow_controls.csv', 'pycs_synthetic_backtest_metrics.csv', 'pycs_return_math_checks.csv', 'pycs_common_mistakes.csv', 'pycs_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. HTML, dependency, and correction checks

```python
print(data['pycs_html_audit.csv'].to_string(index=False))
print(data['pycs_dependency_scope.csv'].to_string(index=False))
print(data['pycs_parameter_corrections.csv'].to_string(index=False))
rows = data['pycs_parameter_corrections.csv'].set_index('item')
assert int(rows.loc['raw_rows','additive_calendar_check']) == 783
```

## 2. Workflow and backtest checks

```python
print(data['pycs_workflow_controls.csv'].to_string(index=False))
print(data['pycs_synthetic_backtest_metrics.csv'].to_string(index=False))
print(data['pycs_return_math_checks.csv'].to_string(index=False))
ret = data['pycs_return_math_checks.csv'].set_index('check')
assert float(ret.loc['absolute_difference','value']) < 1e-8
```

## 3. Mistake controls

```python
print(data['pycs_common_mistakes.csv'].to_string(index=False))
print(data['pycs_validation_controls.csv'].to_string(index=False))
assert data['pycs_common_mistakes.csv'].shape[0] >= 6
```
