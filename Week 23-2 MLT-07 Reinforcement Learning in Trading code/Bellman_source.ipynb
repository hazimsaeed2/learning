# Bellman Equation for Q-Learning

The Bellman equation forms the foundation of Q-learning:

$$Q(s_t, a_t^i) = R(s_t, a_t^i) + \gamma \max[Q(s_{t+1}, a_{t+1})]$$

Where:

- $Q(s_t, a_t^i)$ represents the expected total reward of taking action $a_t^i$ in state $s_t$ at time $t$, then following the optimal policy thereafter.

- $R(s_t, a_t^i)$ is the immediate reward received from taking action $a_t^i$ in state $s_t$.

- $\gamma$ is the discount factor (between 0 and 1) that determines the importance of future rewards. A value closer to 0 makes the agent focus more on immediate rewards, while a value closer to 1 makes it consider future rewards more heavily.

- $\max[Q(s_{t+1}, a_{t+1})]$ represents the maximum expected future reward available from the next state $s_{t+1}$ by taking the best possible action $a_{t+1}$.

In Q-learning, we update our estimates iteratively using:

$$Q(s_t, a_t^i) \leftarrow Q(s_t, a_t^i) + \alpha [R(s_t, a_t^i) + \gamma \max[Q(s_{t+1}, a_{t+1})] - Q(s_t, a_t^i)]$$

Where $\alpha$ is the learning rate determining how much we update our Q-value based on new information.

In our Q-learning implementation of the Marshmallow Experiment:

- **Reward structure**:
  - If a child eats the marshmallow immediately or any time before 7 minutes, they get 1 marshmallow (reward = 1)
  - If they wait the full 7 minutes, they get 2 marshmallows (reward = 2)

- **Parameters**:
  - Discount factor γ = 0.9 (future rewards are discounted by 10% per time step)
  - After waiting 7 minutes, reward = 2

### Interpretation

This table reveals the optimal policy:
- At every time step, the "Wait" action has a higher Q-value than the "Eat" action
- The Q-value for waiting increases as you get closer to the end (time = 1)
- At time = 7 (the beginning), waiting has a Q-value of 2.0, which equals the final reward
- Eating always has a constant Q-value of 1.0 (the immediate reward)

The Q-values for "Wait" are calculated using the Bellman equation, with future rewards discounted by γ = 0.9 at each step. This demonstrates why a rational agent should choose to wait at every step if they can delay gratification.
