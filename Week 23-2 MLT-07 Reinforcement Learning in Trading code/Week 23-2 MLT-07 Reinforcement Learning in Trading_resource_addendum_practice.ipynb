# MLT-07 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for overfitting, fat tails, Bellman/Q-learning, reward design, and DQN controls.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['mlt07_source_pdf_inventory.csv','mlt07_source_zip_inventory.csv','mlt07_data_inventory.csv','mlt07_dependency_fallback_scope.csv','mlt07_overfit_fattail_metrics.csv','mlt07_bellman_marshmallow_table.csv','mlt07_qlearning_results.csv','mlt07_reward_function_reference.csv','mlt07_dqn_config_controls.csv','mlt07_validation_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Sources, data, and dependency scope

```python
print(data['mlt07_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['mlt07_source_zip_inventory.csv'].groupby(['zip','role']).size().to_string())
print(data['mlt07_data_inventory.csv'].to_string(index=False))
print(data['mlt07_dependency_fallback_scope.csv'].to_string(index=False))
assert data['mlt07_data_inventory.csv']['growth_x'].iloc[0] > 100
```

## 2. ML/RL checks

```python
print(data['mlt07_overfit_fattail_metrics.csv'].to_string(index=False))
print(data['mlt07_bellman_marshmallow_table.csv'].to_string(index=False))
print(data['mlt07_qlearning_results.csv'].to_string(index=False))
assert data['mlt07_qlearning_results.csv'].query("test == 'real_market_oos'")['buy_hold_total_pct'].iloc[0] > 300
```

## 3. Reward design and controls

```python
print(data['mlt07_reward_function_reference.csv'].to_string(index=False))
print(data['mlt07_dqn_config_controls.csv'].to_string(index=False))
print(data['mlt07_validation_controls.csv'].to_string(index=False))
assert data['mlt07_validation_controls.csv']['control'].str.contains('Reward design').any()
```
