## Data Extraction and Merging

This section handles the entire process of decompressing and reading raw options data files.

### Key Steps:
1. **Setup and Dependencies**
   - Import required libraries (`os`, `pandas`, and `py7zr` for `.7z` extraction).
   - Define the file paths for the source `.7z` files and the folder where extracted data will be stored.

2. **File Extraction**
   - Automatically identifies all `.7z` files in the source directory.
   - Uses `py7zr` to extract each archive to the designated folder.

3. **Reading and Aggregating Data**
   - Iterates over all extracted `.txt` and `.csv` files.
   - Reads each file into a DataFrame and appends it to a list.
   - Skips any non-text files and handles exceptions gracefully.
   - Deletes each file after reading to free up disk space (optional cleanup).

4. **Final Merging**
   - All monthly DataFrames are concatenated into a single DataFrame `options_data_raw`.
   - Displays a confirmation message once data loading is complete.

This results in a consolidated dataset covering SPX options across multiple months.


```python
import os
import pandas as pd
import py7zr  # ✅ Using py7zr for .7z extraction

# Define paths
path = r"C:\Users\varun\Downloads\EPAT New Lecture\data\\"
extract_path = r"C:\Users\varun\Downloads\EPAT New Lecture\data\extracted\\"

# Ensure extraction folder exists
if not os.path.exists(extract_path):
    os.makedirs(extract_path)

# Get all .7z files
dir_list = [f for f in os.listdir(path) if f.endswith(".7z")]

# Extract .7z files
for file_name in dir_list:
    zip_path = os.path.join(path, file_name)

    with py7zr.SevenZipFile(zip_path, mode='r') as archive:
        archive.extractall(extract_path)

# Read extracted files into DataFrame
dfs = []

for file in os.listdir(extract_path):
    file_path = os.path.join(extract_path, file)

    try:
        if file.endswith(".txt") or file.endswith(".csv"):  # Reads .txt and .csv files
            print(f"Reading: {file_path}")
            monthly_data = pd.read_csv(file_path, sep=',', engine='python')

            if not monthly_data.empty:
                dfs.append(monthly_data)
        else:
            print(f"Skipping non-text file: {file}")
    except Exception as e:
        print(f"Error reading {file}: {e}")

    os.remove(file_path)  # Optional cleanup

# Merge all DataFrames
if dfs:
    options_data_raw = pd.concat(dfs, ignore_index=True)
    print("Data extraction and merging complete!")
else:
    print("No valid data found.")
```

## Cleaning Column Names

The raw dataset contains column names enclosed in square brackets, with inconsistent formatting that may make downstream processing inconvenient. This section standardizes the column names for ease of use.

### Steps:
- Remove square brackets (`[]`) surrounding the names.
- Remove single quotes, if present.
- Replace spaces with underscores to ensure compatibility with typical Python usage.
- Remove any leading underscores that may result from previous replacements.
- Convert all column names to lowercase for consistency.

The result is a set of clean, uniform column names that can be referenced easily in analysis or model-building code.


```python
options_data_raw.columns = (
    options_data_raw.columns
    .str.strip("[]")  # Remove square brackets
    .str.replace("'", "", regex=False)  # Remove single quotes
    .str.replace(" ", "_", regex=False)  # Replace spaces with underscores
    .str.replace(r"^\_+", "", regex=True)  # Remove leading underscores
    .str.replace(r"\[", "", regex=True)  # Remove remaining square brackets
    .str.lower()  # Convert to lowercase
)

# Display the cleaned column names
print(options_data_raw.columns)
```

## Converting Columns to Appropriate Data Types

After inspecting the data, several columns containing numeric values (like implied volatility, Greeks, and prices) were stored as strings. Similarly, date columns were also read as strings. This section converts them to proper numeric and datetime types.

### Steps:
1. **Identify Columns**
   - `float_cols`: All columns that start with `c_` or `p_`, excluding `c_size` and `p_size`, are assumed to be numeric (e.g., delta, gamma, vega).
   - `date_cols`: Any column name containing the substring `"date"` is treated as a date column.

2. **Convert Data Types**
   - Each identified numeric column is converted to `float` using `pd.to_numeric()` with `errors='coerce'` to handle invalid values gracefully.
   - Each identified date column is converted to `datetime` using `pd.to_datetime()`.

3. **Set Index**
   - The `quote_date` column, now in datetime format, is used as the DataFrame index for time-series operations and filtering.

This ensures that the dataset is ready for numerical analysis, time-based grouping, and modeling tasks.


```python
# Identify columns based on their naming patterns
float_cols = [col for col in options_data_raw.columns if col.startswith(("c_", "p_")) and col not in ("c_size", "p_size")]
date_cols = [col for col in options_data_raw.columns if "date" in col.lower()]

# Convert numeric columns
for col in float_cols:
    options_data_raw[col] = pd.to_numeric(options_data_raw[col], errors='coerce')

# Convert date columns
for col in date_cols:
    options_data_raw[col] = pd.to_datetime(options_data_raw[col], errors='coerce')

print("Converted numeric columns:", float_cols)
print("Converted date columns:", date_cols)

# Use quote_date as index
options_data_raw.index = pd.to_datetime(options_data_raw.quote_date)
```

## Classifying Expiry Type

To analyze option behavior across different expiry categories, we define a custom function to classify each option's expiry into types such as Weekly, Monthly, LEAPS, etc.

### Function: `get_expiry_type`
This function compares the quote date and expiry date to assign one of several expiry labels based on calendar logic:

#### Classification Logic:
- **LEAPS**: If the time to expiry is 365 days or more.
- **Near Month / Next Month / Far Month**: If the expiry is the third Friday of the current, next, or the month after next.
- **Monthly (Later Month)**: A third-Friday expiry that is not within the first three months.
- **Weekly**: Any Friday expiry that is not the third Friday of a month.
- **End-of-Month**: If the expiry falls on the last trading day of the month.
- **Daily**: All other expiries (non-Friday, short-dated contracts).

The expiry type is computed row-wise using `.apply()`, where the index (`row.name`) provides the quote date, and the column `expire_date` gives the expiry.

This classification is stored in a new column called `expiry_type` and is essential for analyzing the impact of expiry types on option pricing, liquidity, and strategies.


```python
# Function to classify expiry type
def get_expiry_type(quote_date, expiry_date):
    quote_date = pd.to_datetime(quote_date)
    expiry_date = pd.to_datetime(expiry_date)

    # Check time to expiry
    days_to_expiry = (expiry_date - quote_date).days

    if days_to_expiry >= 365:
        return "LEAPS"

    # Check if it's a Friday
    is_friday = expiry_date.weekday() == 4  # 4 means Friday

    # Compute the third Friday of the current, next, and far month
    third_friday_current = (quote_date + pd.offsets.MonthBegin(1) -
                            pd.offsets.Week(2, weekday=4)).date()
    third_friday_next = (quote_date + pd.DateOffset(months=1) +
                         pd.offsets.MonthBegin(1) -
                         pd.offsets.Week(2, weekday=4)).date()
    third_friday_far = (quote_date + pd.DateOffset(months=2) +
                        pd.offsets.MonthBegin(1) -
                        pd.offsets.Week(2, weekday=4)).date()

    # Monthly Expiry Classification
    if is_friday:
        if expiry_date.date() == third_friday_current:
            return "Near Month"
        elif expiry_date.date() == third_friday_next:
            return "Next Month"
        elif expiry_date.date() == third_friday_far:
            return "Far Month"
        else:
            return "Monthly (Later Month)"

    # Weekly Expiry: Any other Friday that is NOT the third Friday
    if is_friday:
        return "Weekly"

    # End of Month Expiry: Last trading day of the month
    last_trading_day = expiry_date + pd.offsets.MonthEnd(0)
    if expiry_date == last_trading_day:
        return "End-of-Month"

    # Default to daily expiry
    return "Daily"


options_data_raw['expire_date'] = pd.to_datetime(
    options_data_raw['expire_date'])

options_data_raw["expiry_type"] = options_data_raw.apply(
    lambda row: get_expiry_type(row.name, row["expire_date"]),
    axis=1)  # row.name represents the index of the DataFrame
```

## Data processing

## 1. Filter the contracts with last traded value > 0

An option with a last price of zero isn’t realistic for active contracts. Keeping such data could lead to incorrect calculations, such as implied volatility, spreads, or pricing models.

```python
options_data_raw = options_data_raw[(options_data_raw["c_last"] > 0.0)
                                    & (options_data_raw["p_last"] > 0.0)]
options_data_raw
```

## 2. Filter the contracts with strike distance <=0.05

Working with fewer, more meaningful strike prices speeds up calculation. As an example, we are using a strike distance of 5%.

## 3. Remove illiquid contracts bid or ask = 0 X 0

```python
# Filter out options where the call ("c_size") or put ("p_size") has a size of '0 x 0'
# This removes contracts with no market depth (no available bids or asks), indicating illiquidity
options_data_raw = options_data_raw[(options_data_raw['c_size'] != '0 x 0')
                                    & (options_data_raw['p_size'] != '0 x 0')]

# Display the filtered options data
options_data_raw
```

## 4. Filter based on IV values

```python
# Remove rows where either the call option's implied volatility ("c_iv") or the put option's implied volatility ("p_iv") is missing (NaN)
# This ensures we only work with options that have valid implied volatility values
options_data_raw = options_data_raw.dropna(subset=["c_iv", "p_iv"])

# Display the filtered options data
options_data_raw
```

## 5. Filter based on bid and ask values of calls and puts (remove ones with 0 bids or asks)

```python
# Remove options where either the call or put bid/ask prices are zero
# This ensures that only actively traded options with valid market quotes are kept
options_data_raw = options_data_raw[(options_data_raw.c_bid != 0.0) 
                                    & (options_data_raw.p_bid != 0.0) 
                                    & (options_data_raw.c_ask != 0.0) 
                                    & (options_data_raw.p_ask != 0.0)]

# Display the filtered options data
options_data_raw
```
