# OTS-06 resource addendum

This additive notebook preserves the original notebook and adds source-backed validation tables for option-chain quality, calendar tagging, strategy PnL, and backtest controls.

```python
import pandas as pd
from pathlib import Path
base = Path('.')
files = ['ots06_source_pdf_inventory.csv','ots06_source_zip_inventory.csv','ots06_guided_project_inventory.csv','ots06_data_inventory.csv','ots06_option_chain_quality.csv','ots06_trading_day_calendar.csv','ots06_strategy_performance.csv','ots06_sltp_exit_breakdown.csv','ots06_backtest_controls.csv']
data = {f: pd.read_csv(base / f) for f in files}
{k: v.shape for k, v in data.items()}
```

## 1. Source and data inventory

The local monthly SPX option-chain file is the runnable subset of the larger source packet.

```python
print(data['ots06_source_pdf_inventory.csv'][['file','pages','keyword_hits']].to_string(index=False))
print(data['ots06_source_zip_inventory.csv'].groupby('role').size().to_string())
print(data['ots06_data_inventory.csv'].to_string(index=False))
print(data['ots06_option_chain_quality.csv'].to_string(index=False))
assert data['ots06_data_inventory.csv']['rows'].iloc[0] > 0
```

## 2. Calendar and strategy performance

Entries are on the first quote day of each month and exits are tagged separately by calendar rule or SL/TP rule.

```python
print(data['ots06_trading_day_calendar.csv'].to_string(index=False))
print(data['ots06_strategy_performance.csv'].to_string(index=False))
print(data['ots06_sltp_exit_breakdown.csv'].to_string(index=False))
assert data['ots06_strategy_performance.csv']['strategy'].str.contains('long_call').any()
```

## 3. Backtest controls

The controls clarify assumptions that should become explicit parameters in any production options backtester.

```python
print(data['ots06_backtest_controls.csv'].to_string(index=False))
assert data['ots06_backtest_controls.csv']['topic'].str.contains('Costs').any()
```
