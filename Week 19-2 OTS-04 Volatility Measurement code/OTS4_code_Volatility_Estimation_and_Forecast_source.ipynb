### Volatility as the standard deviation of log returns

- **Definition:** Volatility in financial markets is characterized as the standard deviation of log returns.

- **Calculation:** It involves computing the standard deviation of logarithmic returns, which are obtained by taking the natural logarithm of the ratio of current prices to previous prices.

- **Log Returns:** Logarithmic returns offer a more symmetric and convenient representation of price changes compared to simple percentage returns.

- **Indicator of Variation:** The standard deviation of log returns serves as a statistical measure representing the degree of variation or dispersion in the performance of a financial instrument over a specific period.

- **Market Risk:** Higher volatility indicates larger price swings, reflecting increased market turbulence and unpredictability.

- **Risk Assessment:** Investors and traders closely monitor volatility to assess the potential risks associated with an asset, influencing their decision-making processes.

- **Role in Models:** Volatility plays a fundamental role in various financial models, risk management strategies, and option pricing formulas.

- **Market Dynamics:** It provides valuable insights into market dynamics, influencing the pricing and behavior of financial instruments.


```python
# Importing the 'yfinance' library with the alias 'yf' for fetching historical stock price data
import yfinance as yf

# Importing the 'numpy' library with the alias 'np' for numerical and mathematical operations
import numpy as np

# Function to calculate volatility based on historical stock price data


def calculate_volatility(symbol, start_date, end_date):
    # Fetching historical stock price data using the 'yfinance' library
    stock_data = yf.download(symbol, start=start_date, end=end_date)

    # Calculating daily log returns by taking the natural logarithm of the ratio of current and previous day's adjusted close prices
    stock_data['Log_Returns'] = np.log(
        stock_data['Adj Close'] / stock_data['Adj Close'].shift(1))

    # Calculating volatility as the standard deviation of log returns
    volatility = np.std(stock_data['Log_Returns'])

    # Returning the calculated volatility
    return volatility
```

The provided Python code exemplifies an application scenario for calculating and presenting the volatility of a stock, with Apple Inc. (AAPL) chosen as the illustrative example. The `symbol` variable is assigned the value 'AAPL', representing the stock symbol for Apple Inc. Following this, historical stock price data is retrieved from Yahoo Finance using the `yfinance` library, covering the time span from January 1, 2023, to January 1, 2024.

The core calculation is performed by the `calculate_volatility` function, which processes the fetched historical data to determine the volatility of the stock. Specifically, the function computes the standard deviation of daily log returns, providing a statistical measure of the stock's price variation over the specified time frame.

The subsequent part of the code focuses on presenting the computed volatility. The `print` statements are structured to provide a clear and informative output. The printed results include a heading indicating the stock symbol (AAPL) and the date range (January 1, 2023, to January 1, 2024), followed by the calculated volatility expressed as a percentage rounded to three decimal places. This code snippet serves as a practical example of using the `calculate_volatility` function for assessing and communicating the volatility of a specific stock, thereby facilitating financial analysis and decision-making.

```python
# Example usage
# Defining the stock symbol, for example, 'AAPL' represents Apple Inc.
symbol = 'AAPL'
start_date = '2023-01-01'  # Defining the start date for fetching historical data
end_date = '2024-01-01'  # Defining the end date for fetching historical data

# Calculate volatility using the 'calculate_volatility' function with the specified parameters
volatility = calculate_volatility(symbol, start_date, end_date)

# Printing the volatility calculation results with a heading
print(f"Volatility Calculation for {symbol} from {start_date} to {end_date}")
print(f"Volatility: {volatility * 100:.3f}%")
```

## Volatility Estimation using the Parkinson's estimator

Volatility estimation is a crucial aspect of financial analysis, providing insights into the level of price fluctuation in the market. Parkinson's Estimator is one of the methods employed to calculate volatility, offering a straightforward approach to capture the high and low prices within a given time period. This estimator is particularly effective in measuring the inherent volatility of financial assets, contributing to risk assessment and decision-making processes.

## Understanding Parkinson's Estimator

Parkinson's Estimator is based on the high and low prices observed in a time series, utilizing the formula:

$$ \text{Parkinson Estimator Volatility} (σ) = \sqrt{ \frac{1}{4 N ln(2)}\sum log (\frac{High}{Low})^2 } $$

where:
- σ represents the volatility.
- H denotes the highest price within the time period.
- L denotes the lowest price within the time period.

This formula accounts for the logarithmic returns of high and low prices, providing a reliable measure of market volatility.

## Implementation in Python

To apply Parkinson's Estimator in Python, one can fetch historical stock prices using libraries such as 'yfinance' and then calculate volatility using the mentioned formula. This approach offers a simple yet effective means of estimating volatility based on observed high and low prices.

```python
def calculate_parkinson_volatility(symbol, start_date, end_date):
    # Fetch historical stock price data
    stock_data = yf.download(symbol, start=start_date, end=end_date)

    # Calculate Parkinson's historical volatility using the provided formula
    high_prices = stock_data['High']
    low_prices = stock_data['Low']

    log_hl_ratio = np.log(high_prices / low_prices)
    N = len(stock_data)  # Number of observations
    parkinson_volatility = np.sqrt(
        (1 / (4 * N * np.log(2))) * (log_hl_ratio**2).sum())

    return parkinson_volatility


# Example usage
symbol = 'AAPL'  # Example: Apple Inc.
start_date = '2023-01-01'
end_date = '2024-01-01'

# Calculate and print Parkinson's historical volatility with a heading
parkinson_volatility = calculate_parkinson_volatility(
    symbol, start_date, end_date)
print(
    f"Parkinson's Historical Volatility Calculation for {symbol} from {start_date} to {end_date}")
print(f"Volatility: {parkinson_volatility * 100:.3f}%")
```

## Volatility Forecast using EWMA method

```python
def calculate_ewma_volatility(symbol, start_date, end_date, lambda_value):
    # Fetch historical stock price data
    stock_data = yf.download(symbol, start=start_date, end=end_date)

    # Calculate daily log returns
    stock_data['Log_Returns'] = np.log(stock_data['Adj Close'] / stock_data['Adj Close'].shift(1))

    # Calculate EWMA volatility
    log_returns = stock_data['Log_Returns'].dropna()

    # Initialize the first volatility estimate
    ewma_volatility = log_returns.var()

    # Iterate through log returns to update volatility estimates
    for r_squared in log_returns**2:
        ewma_volatility = lambda_value * ewma_volatility + (1 - lambda_value) * r_squared

    ewma_volatility = np.sqrt(ewma_volatility)

    return ewma_volatility

# Example usage
symbol = 'AAPL'  # Example: Apple Inc.
start_date = '2023-01-01'
end_date = '2024-01-01'
lambda_value = 0.94  # Adjust the smoothing parameter as needed (e.g., between 0.9 and 0.99)

# Calculate and print EWMA volatility with a heading
ewma_volatility = calculate_ewma_volatility(symbol, start_date, end_date, lambda_value)
print(f"EWMA Volatility Calculation for {symbol} from {start_date} to {end_date}")
print(f"Volatility: {ewma_volatility * 100:.3f}%")
```

## Volatility Forecast using GARCH (1,1)

GARCH (1,1) is specified as:
σ2t = γV + α * r2t-1 + β * σ2t-1


Needs about 1000 data points to give a stable set of
values (four years of daily data).

To not "blow up" the parameters need to sum to one.

y=1-α-ß

* Have alpha somewhere around 0.01 to 0.1.
* Have beta between 0.9 and 1.0.
* Choose gamma so parameters sum to one.

### Forecasting Volatility

```python
def calculate_garch_volatility(symbol, start_date, end_date, alpha, beta):
    # Fetch historical stock price data
    stock_data = yf.download(symbol, start=start_date, end=end_date)

    # Calculate daily log returns
    stock_data['Log_Returns'] = np.log(
        stock_data['Adj Close'] / stock_data['Adj Close'].shift(1))

    # Drop missing values
    log_returns = stock_data['Log_Returns'].dropna()

    # Initialize variables
    sigma_squared = np.var(log_returns)
    gamma = 1 - alpha - beta

    # Iterate through log returns to update volatility estimates using GARCH(1,1) equation
    for rt in log_returns[:-1]:
        sigma_squared = gamma + alpha * rt**2 + beta * sigma_squared

    return np.sqrt(sigma_squared)


# Example usage
symbol = 'AAPL'  # Example: Apple Inc.
start_date = '2023-01-01'
end_date = '2024-01-01'
alpha = 0.07  # Adjust alpha based on your range (e.g., between 0.01 and 0.1)
beta = 0.92  # Adjust beta based on your range (e.g., between 0.9 and 1.0)

# Calculate and print GARCH(1,1) volatility with a heading
garch_volatility = calculate_garch_volatility(
    symbol, start_date, end_date, alpha, beta)
print(
    f"GARCH(1,1) Volatility Calculation for {symbol} from {start_date} to {end_date}")
print(f"Volatility: {garch_volatility * 100:.3f}%")
```
