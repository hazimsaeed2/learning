# Option Pricing Function

The `calculate_option_prices` function is designed to compute the price of options using the Black-Scholes (BS) model. This function takes several parameters such as stock price, strike price, interest rate, days to expiration, volatility, and option type as inputs.

## Syntax
`calculate_option_prices(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type)`

## Parameters

- `stock_price`: The current stock price.
- `strike_price`: The strike price of the option.
- `interest_rate`: The risk-free interest rate.
- `days_to_expiration`: The number of days until the option expiration.
- `volatility`: The volatility of the underlying asset.
- `option_type`: The type of option, either 'call' or 'put'.



```python
# Importing the Black-Scholes (BS) model from the mibian library
from mibian import BS

# Importing the 'warnings' module to filter out warnings
import warnings

# Filtering out warnings to improve code readability
warnings.filterwarnings('ignore')

# Function to calculate call and put prices

def calculate_option_prices(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type):
    # Creating a BS object with provided option parameters
    option = BS([stock_price, strike_price, interest_rate,
                days_to_expiration], volatility=volatility)

    # Checking the option type and calculating the corresponding option price
    if option_type.lower() == 'call':
        price = option.callPrice
    elif option_type.lower() == 'put':
        price = option.putPrice
    else:
        # Raising an error for an invalid option type
        raise ValueError("Invalid option type. Use 'call' or 'put'.")

    # Rounding the calculated option price to 3 decimals
    return round(price, 3)
```

# Option Pricing Example

In this example, we demonstrate the usage of the `calculate_option_prices` function to calculate the prices of call and put options using the Black-Scholes (BS) model.

## Input Parameters

- `stock_price`: $100.0
- `strike_price`: $105.0
- `interest_rate`: 0%
- `days_to_expiration`: 30 days
- `volatility`: 30% (You may need to adjust this value based on market conditions)
- `option_type`: 'call'

```python
# Example usage
stock_price = 100.0
strike_price = 105.0
interest_rate = 0
days_to_expiration = 30
volatility = 30  # You may need to adjust this value based on market conditions
option_type = 'call'

call_price = calculate_option_prices(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type)
print(f"Call Option Price: {call_price}")
```

### Put Option Calculation
Here, we change the `option_type` to `put` and calculate the price of a put option using the same set of input parameters.

Feel free to adjust the input parameters as needed for your specific analysis.

```python
# Put option example
option_type = 'put'
put_price = calculate_option_prices(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type)
print(f"Put Option Price: {put_price}")
```
