# Option Greeks Calculator Function

This function utilizes the Black-Scholes (BS) model from the `mibian` library to calculate and print option Greeks for a given set of parameters. The option Greeks include Delta, Theta, Rho, and Gamma.

## Parameters:
- `stock_price`: Current stock price.
- `strike_price`: Strike price of the option.
- `interest_rate`: Annual interest rate.
- `days_to_expiration`: Number of days until the option expiration.
- `volatility`: Volatility of the underlying asset.
- `option_type`: Type of option, either 'call' or 'put'.

## Option Greeks:
- **Delta**: Measures the sensitivity of the option price to changes in the underlying stock price.
- **Theta**: Represents the time decay of the option price.
- **Rho**: Measures the sensitivity of the option price to changes in interest rates.
- **Gamma**: Represents the rate of change of Delta with respect to changes in the underlying stock price.

````python
# Importing the Black-Scholes (BS) model from the mibian library
from mibian import BS

# Importing the 'warnings' module to filter out warnings
import warnings

# Filtering out warnings to improve code readability
warnings.filterwarnings('ignore')

# Function to calculate and print option Greeks
def calculate_option_greeks(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type):
    option = BS([stock_price, strike_price, interest_rate,
                days_to_expiration], volatility=volatility)

    if option_type.lower() == 'call':
        delta = option.callDelta
        theta = option.callTheta
        rho = option.callRho
        gamma = option.gamma
    elif option_type.lower() == 'put':
        delta = option.putDelta
        theta = option.putTheta
        rho = option.putRho
        gamma = option.gamma
    else:
        raise ValueError("Invalid option type. Use 'call' or 'put'.")

    # Printing option Greeks rounded to 3 decimals
    print(f"{option_type.capitalize()} Option Greeks:")
    print(f"Delta: {round(delta, 3)}")
    print(f"Theta: {round(theta, 3)}")
    print(f"Rho: {round(rho, 3)}")
    print(f"Gamma: {round(gamma, 3)}")
````

## Usage Example:
```python
calculate_option_greeks(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type)


````python
# Setting up initial parameters
stock_price = 100.0           # Current stock price
strike_price = 105.0          # Strike price of the option
interest_rate = 0             # Annual interest rate
days_to_expiration = 30       # Number of days until the option expiration
volatility = 30               # Volatility of the underlying asset (You may need to adjust this value based on market conditions)
````

````python
# Set the type of option 
option_type = 'call'

# Call the function to calculate option Greeks
calculate_option_greeks(stock_price, strike_price, interest_rate, days_to_expiration, volatility, option_type)
````
